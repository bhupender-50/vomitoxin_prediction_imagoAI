```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.impute import SimpleImputer
from xgboost import XGBRegressor
import optuna
import joblib
import logging
```

```python
# Step 3: Improve Data Quality
# Handle NaN values in the target variable (vomitoxin_ppb)
if data['vomitoxin_ppb'].isna().sum() > 0:
    logging.info("Handling NaN values in vomitoxin_ppb...")
    data = data.dropna(subset=['vomitoxin_ppb'])  # Remove rows with NaN in target

# Remove outliers in the target variable 
mean_vomitoxin = data['vomitoxin_ppb'].mean()
std_vomitoxin = data['vomitoxin_ppb'].std()
data = data[(data['vomitoxin_ppb'] >= mean_vomitoxin - 3 * std_vomitoxin) & (data['vomitoxin_ppb'] <= mean_vomitoxin + 3 * std_vomitoxin)]
logging.info("Outliers removed.")
```

```python
# Step 4: Preprocess the Data
# Drop the 'hsi_id' column
data = data.drop(columns=['hsi_id'])

# Separate features (X) and target (y)
X = data.drop(columns=['vomitoxin_ppb'])
y = data['vomitoxin_ppb']

# Handle NaN and infinite values in features (X)
imputer = SimpleImputer(strategy='mean')
X = pd.DataFrame(imputer.fit_transform(X), columns=X.columns) 
X = np.nan_to_num(X, nan=np.nan, posinf=1e10, neginf=-1e10)  

# Normalize the spectral data (columns 0 to 447)


logging.info("Normalizing spectral data...")
scaler = StandardScaler()
spectral_features = data.drop(columns=['vomitoxin_ppb'])
scaled_features = scaler.fit_transform(spectral_features)



# Save the scaler for later use
joblib.dump(scaler, "scaler.pkl")
logging.info("Scaler saved.")

# Combine scaled features with the target variable
processed_data = pd.DataFrame(scaled_features, columns=spectral_features.columns)
processed_data['vomitoxin_ppb'] = data['vomitoxin_ppb']

# Visualize average reflectance over wavelengths
plt.figure(figsize=(10, 6))
plt.plot(spectral_features.columns.astype(float), processed_data.mean()[:-1])
plt.title("Average Reflectance Over Wavelengths")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Reflectance")
plt.show()
```

```python
# Step 5: Feature Engineering
# Perform feature selection (select top 20 features based on correlation with target)

logging.info("Performing feature selection...")
selector = SelectKBest(score_func=f_regression, k=20)
X_selected = selector.fit_transform(scaled_features, data['vomitoxin_ppb'])

# Get the selected feature names
selected_feature_names = spectral_features.columns[selector.get_support()]
print("Selected features:", selected_feature_names)
```

```python
# Step 6: Split the Data
logging.info("Splitting data into training and testing sets...")
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)
```

```python
# Step 7: Experiment with Advanced Models
# Random Forest Regressor
logging.info("Training Random Forest model...")
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
try:
    rf_scores = cross_val_score(rf_model, X_selected, y, scoring='neg_mean_absolute_error', cv=5, error_score='raise')
    print("Random Forest MAE:", -rf_scores.mean())
except Exception as e:
    print("Error during Random Forest training:", e)

# XGBoost Regressor
logging.info("Training XGBoost model...")
xgb_model = XGBRegressor(n_estimators=200, learning_rate=0.1, max_depth=5, random_state=42)
try:
    xgb_scores = cross_val_score(xgb_model, X_selected, y, scoring='neg_mean_absolute_error', cv=5, error_score='raise')
    print("XGBoost MAE:", -xgb_scores.mean())
except Exception as e:
    print("Error during XGBoost training:", e)
```

```python
# Step 8: Train the Final Model (XGBoost)
logging.info("Training the final XGBoost model...")
final_model = XGBRegressor(n_estimators=200, learning_rate=0.1, max_depth=5, random_state=42)
final_model.fit(X_train, y_train)

# Save the final model
joblib.dump(final_model, "final_vomitoxin_model.pkl")
logging.info("Final model saved.")
```

```python
# Step 9: Evaluate the Final Model
logging.info("Evaluating the final model...")
y_pred_final = final_model.predict(X_test)

# Calculate evaluation metrics
mae_final = mean_absolute_error(y_test, y_pred_final)
rmse_final = np.sqrt(mean_squared_error(y_test, y_pred_final))

print(f"Final MAE: {mae_final}")
print(f"Final RMSE: {rmse_final}")

# Plot actual vs predicted values
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred_final)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.title("Actual vs Predicted Vomitoxin (ppb)")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.show()

# Residual analysis
residuals = y_test - y_pred_final
plt.figure(figsize=(8, 6))
sns.histplot(residuals, kde=True)
plt.title("Residual Distribution")
plt.xlabel("Residuals")
plt.show()
```
